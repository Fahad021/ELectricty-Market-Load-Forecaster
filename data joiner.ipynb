```python
import pandas as pd
df_edmonton       = pd.read_csv('edmonton,AB.csv').iloc[:, 1:]
df_calgary        = pd.read_csv('calgary,AB.csv').iloc[:, 1:]
df_fortmcmurray   = pd.read_csv('fort mcmurray,AB.csv').iloc[:, 1:]
df_lethbridge     = pd.read_csv('lethbridge,AB.csv').iloc[:, 1:]
df_medicinehat    = pd.read_csv('medicine hat,AB.csv').iloc[:, 1:]
df_slavelake      = pd.read_csv('red deer,AB.csv').iloc[:, 1:]
df_reddeer        = pd.read_csv('red deer,AB.csv').iloc[:, 1:]
```

```python
df_edmonton.columns      = [str(col) + '_'+ 'edmonton' for col in df_edmonton.columns]
df_calgary.columns       = [str(col) + '_'+ 'calgary'      for col in df_calgary.columns]
df_fortmcmurray.columns  = [str(col) + '_'+ 'fortmcmurray' for col in df_fortmcmurray.columns]
df_lethbridge .columns   = [str(col) + '_'+ 'lethbridge'   for col in df_lethbridge .columns]
df_medicinehat.columns   = [str(col) + '_'+ 'medicinehat'  for col in df_medicinehat.columns]
df_slavelake .columns    = [str(col) + '_'+ 'slavelake'    for col in df_slavelake.columns]
df_reddeer.columns       = [str(col) + '_'+ 'reddeer'      for col in df_reddeer.columns]
```

```python
df_all = pd.concat([df_edmonton, df_calgary, df_fortmcmurray,
                    df_lethbridge, df_medicinehat, df_slavelake , df_reddeer], axis=1)
df_all.dtypes
```

```python
df_all.to_csv('wwo_histrical_db.csv')
```
